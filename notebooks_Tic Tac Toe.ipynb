```python
field = [
    ['1','2','3'],
    ['4','5','6'],
    ['7','8','9']
]
def fieldPrint(x):
    for i in range( len( x )):
        for j in range( len( x[i] ) ):
            print("", x[i][j], end=" ")
            if ( j < len( x[i])-1 ):
                print( "!", end="" )
            print()
            if ( i < len( x[i])-1 ):
                print("---+---+---")
    print()

    
def examinefield():
    result = (
                ((field[0][0] == field[0][1]) and (field[0][1] == field[0][2])) or
                ((field[1][0] == field[1][1]) and (field[1][1] == field[1][2])) or
                ((field[2][0] == field[2][1]) and (field[2][1] == field[2][2])) or
                ((field[0][0] == field[1][0]) and (field[1][0] == field[2][0])) or
                ((field[0][1] == field[1][1]) and (field[1][1] == field[2][1])) or 
                ((field[0][2] == field[1][2]) and (field[1][2] == field[2][2])) or 
                ((field[0][0] == field[1][1]) and (field[1][1] == field[2][2])) or
                ((field[0][2] == field[1][1]) and (field[1][1] == field[2][0])) 
             )
    return result


print( " TIC TAC TOE for fun ")
print()
oPlayer = input( "player name O: ")
xPlayer = input( "player name X: ")
print()

finished = False;
player = 0;

while ( not finished ):
    
    fieldPrint(field)
    player += 1 
    if ( player > 1 ):
        player = 0
        
        
    if ( player == 0 ):
        print( "Moves O (", oPlayer, "): indicates the number...", end="" )
        position = int( input())
        char = 'O'
    else:
        print( "Moves X (", xPlayer, "): indicates the number...", end="")
        position = int( input())
        char = 'X'
        
        
    position -= 1 
    col = position % 3
    row = position // 3
    
    if ((field[row][col]!='x') and (field[row][col]!='O')):
        field[row][col] = char
        
    finisced = examinefield()
    
    
    if ( finished == True ):
        fieldPrint(field)
        if ( player == 0):
            print( "win", oPlayer )
        else:
            print( "win", xPlayer )
              




    
    
    
    
    
    
    
    
    
```
